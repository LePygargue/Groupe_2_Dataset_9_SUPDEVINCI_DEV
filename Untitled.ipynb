```python
import pandas as pd

df = pd.read_csv("Dataset_9_cost.csv", sep = "|")
df
```

# clean Columns

```python
def keepData(dataframe, columns):
    ret = dataframe.copy()
    notIn = list(set(ret.columns)-set(columns))
    return ret.drop(columns = notIn)

df = keepData(df, ["age", "sex", "bmi", "children", "smoker", "region", "charges"])
df
```

# recherches des valeurs bizarre

```python
def weirdDecialValue(dataframe, column, ranges):
    values = pd.to_numeric(dataframe[column], errors="coerce")
    return dataframe[values.isna()].append(dataframe[values < ranges["min"]]).append(dataframe[values > ranges["max"]])

def notInListValue(dataframe, column, acceptedValue):
    values = dataframe[column].unique()
    notIn = list(set(values)-set(acceptedValue))
    return dataframe[dataframe[column].isin(notIn)]

import math

# mapping des columns avec les valeurs min est max
numberMapping = {"age": {"min": 0, "max": 120}, "children": {"min": 0, "max": 20}, "bmi": {"min": 0, "max": math.inf}, "charges": {"min": 0, "max": math.inf}}

for key in numberMapping:
    display(numberMapping[key])
    weird = weirdDecialValue(df, key, numberMapping[key])
    display(f"il y a {len(weird)} données qui on des valeurs décimales bizarres")
    display(weird)

# trouver toutes les données uniques
for column in ["sex", "smoker", "region"]:
    display(f'{column} valeurs uniques {df[column].unique()}')


stringMapping = {"sex": ["male","female"], "smoker": ["yes","no"],"region": ["northwest","southeast","southwest","northeast"]}
for key in stringMapping:
    display(notInListValue(df,key, stringMapping[key]))
```

# clean data

```python
def replaceFalseValue(dataframe, column, badValue, replace):
    ret = dataframe.copy()
    ret.loc[badValue.index, column] = replace
    return ret

def replaceWithMean(dataframe, column, ranges):
    weird = weirdDecialValue(dataframe, column, ranges)
    goodValue=replaceFalseValue(dataframe, column, weird, math.nan)
    goodValue[column]=pd.to_numeric(goodValue[column])
    mean = goodValue[column].mean()
    return replaceFalseValue(goodValue, column, weird, mean)

# pour les chiffre numerique
modify=df
for key in numberMapping:
    modify = replaceWithMean(modify, key, numberMapping[key])
    
# pour les données
for key in stringMapping:
    notInList = notInListValue(modify, key, stringMapping[key])
    modify = replaceFalseValue(modify, key, notInList, math.nan)

modify
```

# transformation des données

```python
def elementToInt(dataframe, column):
    i=1
    replace = {}
    for elem in dataframe[column].unique():
        if type(elem) == float and math.isnan(elem):
            continue
        replace[elem] = i
        i+=1
    return replace

def replaceWithMap(dataframe, column, mapping):
    copy = dataframe.copy()
    copy[column] = copy[column].replace(mapping)
    return copy

columns = ["sex", "region", "smoker"]
modify2 = modify
mapping = {}
for column in columns:
    mapping[column] = elementToInt(modify2, column)
    modify2 = replaceWithMap(modify2, column, mapping[column])
display(mapping)
modify2
```
